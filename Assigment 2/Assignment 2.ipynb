# Encoding

```python
df.head()
```

```python
df = pd.read_excel("insurance.xlsx")
df
```

```python
region1 = df.region.unique()
region1
```

```python
df.region = df.region.replace(['southwest', 'southeast', 'northwest', 'northeast'],[1,2,3,4])
```

```python
smoker1 = df.smoker.unique()
smoker1
```

```python
df.smoker = df.region.replace(['yes', 'no'],[0,1])
```

```python
gender1 = df.gender.unique()
gender1
```

```python
df.gender = df.gender.replace(['female', 'male'],[1,2])
```

```python
df
```

# Label Encoding

```python
df1 = pd.read_excel("insurance.xlsx")
df1
```

```python
df1.smoker = le.fit_transform(df1.smoker)
df1
```

# Using Loop

```python
df2 = pd.read_excel("insurance.xlsx")
df2
```

```python
df2.columns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for col in df2.columns:
    if df2[col].dtype == np.number:
        continue
    else:
        df2[col] = le.fit_transform(df2[col])
```
