# Encoding

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in df.columns:
    if df[col].dtype == np.number:
        continue
    else:
        df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
x = df[['size(cm)']]
```

```python
y = df[['class(y)']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.3,random_state=55)
```

```python
xtrain.head()
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c
```

```python
reg.predict_proba(xtest) #Probability
```

```python
reg.predict(xtest)
```

```python
ytest.head()
```

```python
Log_score = reg.score(xtest,ytest)
```

```python
Log_score
```

```python
mae1 = mean_absolute_error(ytest,reg.predict(xtest))
```

```python
mae1
```

```python
mse1 = mean_squared_error(ytest,reg.predict(xtest))
```

```python
mse1
```

# Dataframe create

```python
data = {'Name ': ['Logistic_mae','Logistic_mse','Logistic_accuracy'],'Accuracy':[mae1,mse1,Log_score]}
```

```python
df_after_pred = pd.DataFrame(data=data)
```
